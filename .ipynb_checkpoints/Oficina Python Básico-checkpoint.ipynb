####  Adicionando e Removendo elementos
##### Append
O método `append` adiciona à ultima posição da lista o objeto recebido como argumento.

```python
sistemaSolar = ['Mercúrio', 'Vênus', 'Terra', 'Marte', 'Júpiter', 'Saturno']
sistemaSolar.append('Netuno')
print(sistemaSolar)
```

##### Insert
O método `insert` adiciona a qualquer posição da lista, o objeto passado como argumento.

```python
sistemaSolar.insert(6, 'Urano')
print(sistemaSolar)
```

##### Remove 
O método `remove`, remove uma ocorrência do valor passado como arguemento.

```python
sistemaSolar.remove('Saturno')
print(sistemaSolar)
```

##### Pop
O método `pop` remove a última posição da lista e a retorna.

```python
sistemaSolar.append('Plutão')
print(sistemaSolar)
```

```python
print(sistemaSolar)
```

Também podemos usar o método `pop` passando uma posição da lista como argumento, desta forma ele removerá o objeto daquela posição e o retornará.

#### Organizando listas
##### Sort
O método `sort` altera a lista deixando-a em ordem crescente.

```python
sistemaSolar.sort()
print(sistemaSolar)
```

Podemos utilizar o argumetno `reverse = True` para alterar nossa lista para a ordem decrescente.

```python
sistemaSolar.sort(reverse=True)
print(sistemaSolar)
```

##### Sorted
A função `sorted()` recebe como argumento uma lista, a diferença desta função para o método sort é que a função sorted não altera nosso objeto, ao invés disto, ela retorna uma cópia ordenada de nossa lista.

```python
sistemaSolar = ['Mercúrio', 'Vênus', 'Terra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Netuno']
sistemaSolarOrdemAlfabetica = sorted(sistemaSolar)
print(sistemaSolar)
print(sistemaSolarOrdemAlfabetica)

sistemaSolar = ['Mercúrio', 'Vênus', 'Terra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Netuno']
sistemaSolarOrdemAlfabetica = sorted(sistemaSolar, reverse=True)
print(sistemaSolarOrdemAlfabetica)
```

##### Reverse
O método `reverse` inverte a posição dos objetos em nossa lista.

```python
sistemaSolar = ['Mercúrio', 'Vênus', 'Terra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Netuno']

sistemaSolar.sort()
print(sistemaSolar)
sistemaSolar.reverse()
print(sistemaSolar)
sistemaSolar.reverse()
print(sistemaSolar)
sistemaSolar.reverse()
print(sistemaSolar)
```

```python
sistemaSolar = ['Mercúrio', 'Vênus', 'Terra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Netuno']
sistemaSolar.reverse()
print(sistemaSolar)
```

### Copiando listas
Se temos uma lista alocada em uma variável, se atribuírmos esta variável a outra variável as duas variáveis terão o mesmo endereço de memória, ou seja, qualquer modificação feita em qualquer uma das listas terá o mesmo efeito na outra.  
O mesmo endereço de memória aponta para as duas variáveis. Veja o exemplo:

```python
planetas = sistemaSolar
planetas.append('Kepler 22b')
planetas.append('Upsilon Andromedae b')
planetas.append('WASP-11b/HAT-P-10b')
print(planetas)
```

```python
planetas==sistemaSolar
```

Observe que mesmo adicionando novos planetas a lista planetas, a lista sistemaSolar recebe as mesmas modificação.

```python
print(sistemaSolar)
```

`is` verifica se duas variáveis possuem o mesmo endereço de memória.

```python
planetas is sistemaSolar
```

Printando o endereço de memória.

```python
print(id(planetas))
print(id(sistemaSolar))
```

Se não desejamos que as duas variáveis possuam o mesmo endereço de memória, precisamos atribuir a nova lista uma cópia da lista a ser atribuída.

```python
sistemaSolar = ['Mercúrio', 'Vênus', 'Terra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Netuno']
planetas = sistemaSolar[:] # Cria uma cópia da lista sistemaSolar
```

```python
sistemaSolar==planetas
```

```python
planetas.append('Kepler 22b')
planetas.append('Upsilon Andromedae b')
planetas.append('WASP-11b/HAT-P-10b')

print(sistemaSolar)
print(planetas)
```

```python
sistemaSolar==planetas
```

```python
sistemaSolar is planetas
```

Verifique que agora as duas variáveis possuem endereços de memória diferentes

```python
print(id(planetas))
print(id(sistemaSolar))
```

## Controle de fluxo
Nem sempre nosso código deve se comportar de maneira igual para todas as situações, as vezes precisamos controlar o fluxo de nossas operações em situações específicas e para isso podemos utilizar `if` e uma condição.  
No caso da condição de `if` for satisfeita as proximas linhas identadas abaixo dele serão executadas, caso contrário não.
    
### Testando uma condição única.

```python
if 10 < 30:
    print('10 é menor que 30')
```

```python
if 10 > 30:
    print('10 é menor que 30')
```

```python
verdade = True
if verdade:
    print('Verdade')
```

```python
verdade = True

if not verdade:
    print('Verdade')
```

```python
if (7 < 3) | (5 > 1):
    print('Pelo menos um é verdade')
```

Podemos verificar se um valor esta em uma lista utilizando `in`.

```python
generosMusicaisBrasileiros = ['pagode', 'bossa nova', 'mpb', 'samba', 'axe']
generoMusical = 'heavy metal'

if generoMusical in generosMusicaisBrasileiros:
    print(generoMusical.title() + ' é um genero musical brasileiro.')
    
else:
    print(generoMusical.title() + ' não é um genero musical nacional.')
```

```python
generosMusicaisBrasileiros = ['pagode', 'bossa nova', 'mpb', 'samba', 'axe']
generoMusical = 'axe'

if generoMusical in generosMusicaisBrasileiros:
    print(generoMusical.title() + ' é um genero musical brasileiro.')
    
else:
    print(generoMusical.title() + ' não é um genero musical nacional.')
```

Podemos verificar se um valor não esta em uma lista utilizando `not in`.

```python
filmesDeHeroi = ['liga da justica',
                'homem-aranha: no aranhaverso', 
                'incrivel hulk', 
                'vingadores: ultimato', 
                'vingadores: guerra infinita']

filme = 'dumbo'

if filme not in filmesDeHeroi:
    print(filme.title() + ' não é um filme de heroi.')
else:
    print(filme.title() + ' é um filme de heroi.')
```

### Testando várias condições
#### elif
Quando precisamos testar mais de 2 situações e podemos utilizar `elif` para lidar com essas condições.  
Assim que uma condição é satisfeita, ela executa sua função.

```python
idade = int(input('Qual a sua idade?\n\t- '))

if idade < 4:
    print('Você entra de graça!')

elif (idade <= 21) | (idade >= 65):
    print('Você paga meia entrada!')
    
else:
    print('Você paga completa!')
```

```python
idade = int(input('Qual a sua idade?\n\t- '))

if idade < 4:
    print('Você entra de graça!')

elif idade <= 21:
    print('Você paga meia entrada!')
    
elif idade < 65:
    print('Você paga completa!')
    
else:
    print('Você paga meia entrada!')
```
