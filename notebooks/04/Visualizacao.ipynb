## Gráfico de linha

### Apenas com matplot

```python
import matplotlib.pyplot as plt
numeros = [1, 2, 3, 4, 5]
quadrados = [numero**2 for numero in numeros]

plt.plot(quadrados, linewidth = 5)

# Set chart title and label axes.
plt.title('Square Numbers', fontsize = 24)
plt.xlabel('Value', fontsize = 14)
plt.ylabel('Square of Value', fontsize = 14)

# Set size of tick labels.
plt.tick_params(axis = 'both', labelsize = 14)

plt.show()
```

## Gráfico de barras

Iremos criar uma classe chamada de "Die", essa classe imita um dado. Para ser iniciada, devemos passar Dice(Numero de faces no dado). Possui um método chamado de "dice_roll" que simula o lançamento do dado.


```python
import numpy as np

class Die():
    """A class representing a single dice roll."""
    
    def __init__(self, n_sides=6):
        """Assume a six-sided die."""
        self.n_sides = n_sides
        
    def dice_roll(self, n_times=1):
        """Return a random number between 1 and number of sides"""
        return np.random.randint(1, self.n_sides+1, n_times)
```

Aqui iremos simular o lançamento de 2 dados com 5 faces e soma-las e também o lançamento de 1 dado com 10 faces. Iremos fazer 10000 lançamentos e comparar a frequência de resultados obtidos.

```python
import matplotlib.pyplot as plt

d10 = Die(10)
d5 = Die(5)

results_d10 = []

for roll in range(10000):
    results_d10.append(d10.dice_roll())
    

results_2d5 = []

for roll in range(10000):
    results_2d5.append(d5.dice_roll() + d5.dice_roll())

frequencies_d10 = []

for value in range(1, 11):
    frequency = results_d10.count(value)
    frequencies_d10.append(frequency)

frequencies_2d5 = []
for value in range(2, 11):
    frequency = results_2d5.count(value)
    frequencies_2d5.append(frequency)
```

Iremos utilizar o comando plt.bar para dizer ao python para guardar em sua memória 1 gráfico de barras, iremos utilizar o comando duas vezes para os gráficos fiquem sobrepostos e por fim iremos utilizar plt.show() para que o python mostre os gráficos.

### Matplot

```python
frequencies_2d5.insert(0,0)    

plt.bar(range(1,11), frequencies_2d5)
plt.bar(range(1,11), frequencies_d10, alpha = 0.67, color = 'red')

plt.title('2D5 x D10\nRolled 1000 Times')
plt.ylabel('Frequency')
plt.xlabel('Results')
plt.legend(['2D5','D10'])
plt.show()
```

### Seaborn

```python
sns.barplot(x = list(range(1,11)), y = frequencies_2d5, color = "blue")
sns.barplot(x = list(range(1,11)), y = frequencies_d10, color = "red",alpha=0.7)
```
